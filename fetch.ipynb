```python

from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi
from dotenv import load_dotenv
import os
import pandas as pd

load_dotenv(override=True)
USERNAME = os.getenv("USERNAME")
PASSWORD = os.getenv("PASSWORD")
DATABASE_NAME = os.getenv("DATABASE_NAME")
print("USERNAME: ", USERNAME)
print("PASSWORD: ", PASSWORD)
print("DATABASE_NAME: ", DATABASE_NAME)
uri = "mongodb+srv://"+USERNAME+":"+PASSWORD+"@chatui.63yozgy.mongodb.net/?retryWrites=true&w=majority&appName=ChatUI"

# Create a new client and connect to the server
client = MongoClient(uri, server_api=ServerApi('1'))

# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)

# Fetch data from the database according to the collection name
def fetch_data(database, collection_name):
    collection = database[collection_name]
    items = list(collection.find())
    return items

# Get the database
database = client.get_database(DATABASE_NAME)
# Show all the table names
print(database.list_collection_names())

# Fetch data and turn into dataframe
feedback_data = fetch_data(database, "feedback")
feedback_df = pd.DataFrame(feedback_data)
conversation_data = fetch_data(database, "conversations")
conversation_df = pd.DataFrame(conversation_data)

# Merge feedback and conversation data
merge_df = pd.merge(feedback_df, conversation_df, left_on="conversationId", right_on="_id")
print(merge_df)

# # Print conversation of the first feedback according to the conversationId
# first_feedback = feedback_data[0]
# conversation_id = first_feedback["conversationId"]
# conversation = [conversation for conversation in conversation_data if conversation["_id"] == conversation_id][0]
# message = [message for message in conversation["messages"] if message["id"] == first_feedback["messageId"]][0]
# print(message)


```
